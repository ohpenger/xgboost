```python
import numpy as np
import pandas as pd
import xgboost as xgb
import seaborn as sns
```

```python
xgb_train = xgb.DMatrix("../data/mushroom/agaricus.txt.train?format=libsvm")
xgb_test = xgb.DMatrix("../data/mushroom/agaricus.txt.test?format=libsvm")
type(xgb_test)
```

```python
params = {'objective': 'binary:logistic',
          'booster': 'gbtree',
          'eta': 0.1,
          'gamma': 0.1,
          'min_child_weight': 1,
          'max_depth': 6,}
num_round = 24
watchlist = [(xgb_train, 'train'), (xgb_test, 'test')]
model = xgb.train(params, xgb_train, num_round, watchlist)
```

```python
model.save_model('../saved_models/binary_001.model')
```

```python
bst = xgb.Booster(model_file='../saved_models/binary_001.model')
preds = bst.predict(xgb_test)
print(preds)
```

```python
postive = preds[preds > 0.5]
print(postive)
```
